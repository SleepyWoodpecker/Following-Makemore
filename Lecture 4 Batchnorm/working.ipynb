```python
import torch
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n1 = torch.randn(10000000) / 10
n1.max(), n1.min(), n1.std()

n2 = torch.randn(10000000) / 5
n2.max(), n2.min(), n2.std()
```

```python
legends = []
for i, n in enumerate([n1, n2]): 
    hy, hx = torch.histogram(n, density=True)
    plt.plot(hx[:-1].detach(), hy.detach())

    print(f"STD: {n.std()} | MEAN: {n.mean()}")
    legends.append(i)

plt.legend(legends)
```
